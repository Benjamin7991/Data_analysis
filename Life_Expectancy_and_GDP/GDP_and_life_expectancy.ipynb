```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv('all_data.csv')
```

```python
df['Country'] = df.Country.apply(lambda x: 'United States' if x == 'United States of America' else x)
```

```python
df.columns
```

```python
df = df.rename(columns = {'Life expectancy at birth (years)': 'life_expectancy'})
```

```python
df.head()
```

```python
life_per_country = df.groupby('Country')['life_expectancy'].mean().reset_index()
```

```python
ax = plt.subplot()
plt.xticks(rotation = 50)
plt.title('Average Life Expectancy per Country', size=14)
plt.xlabel('Country')
plt.ylabel('Life Expectancy (years)')


plt.bar(life_per_country.Country, life_per_country.life_expectancy)
plt.show()
```

```python
df.head()
```

```python
chile = df.loc[df.Country == 'Chile']
china = df.loc[df.Country == 'China']
germany = df.loc[df.Country == 'Germany']
mex = df.loc[df.Country == 'Mexico']
usa = df.loc[df.Country == 'United States']
zim = df.loc[df.Country == 'Zimbabwe']


plt.figure(figsize= (10,10))
plt.plot(chile.Year, chile.life_expectancy, label='Chile')
plt.plot(china.Year, china.life_expectancy, label='China')
plt.plot(germany.Year, germany.life_expectancy, label='Germany')
plt.plot(mex.Year, mex.life_expectancy, label='Mexico')
plt.plot(usa.Year, usa.life_expectancy, label='USA')
plt.plot(chile.Year, zim.life_expectancy, label='Zimbabwe')
plt.legend()
plt.ylabel('Life Expectancy (years)')
plt.xlabel('Year')
plt.title('Average Life Expectancy Over Time', size = 16)
plt.show()
```

```python
plt.figure(figsize = (10,10))
ax = plt.subplot()
plt.plot(chile.Year, chile.GDP, label = 'Chile')
plt.plot(chile.Year, china.GDP, label = 'China')
plt.plot(chile.Year, germany.GDP, label = 'Germany')
plt.plot(chile.Year, mex.GDP, label = 'Mexico')
plt.plot(chile.Year, usa.GDP, label = 'USA')
plt.plot(chile.Year, zim.GDP, label='Zimbabwe')
plt.legend(loc = 2)
plt.title('GDP over Time', size = 18)
plt.ylabel('GDP (in trillions of US Dollars)', size = 15)
plt.show()
```

```python

plt.figure(figsize = (10, 8))

plt.scatter(usa.GDP, usa.life_expectancy,  label = 'USA')
plt.scatter(mex.GDP, mex.life_expectancy, label = 'Mexico')
plt.scatter(china.GDP, china.life_expectancy,  label = 'China')

plt.title('Life Expectancy vs GDP', size = 18)
plt.xlabel('GDP in Trillions of US Dollars')
plt.ylabel('Life Expectancy (years)')
plt.legend()
plt.show()
```
