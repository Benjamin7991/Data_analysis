```python
import pandas as pd

df = pd.read_csv('hate_crime_data.csv')
```

```python
df.head()
```

```python
income = df.sort_values(by='median_household_income', ascending = True)
```

```python
income.head()
```

```python
import matplotlib.pyplot as plt


N = 50
x = income['avg_hatecrimes_per_100k_fbi']
y = income['median_household_income']

plt.xlabel('Number of Hate Crimes per 100k people')
plt.ylabel('Median Household Income')
plt.title('Correlation Between Household Income and the Number of Hate Crimes Committed in the US')

#area = (30 * np.random.rand(N))**2  # 0 to 15 point radii

plt.scatter(x, y)
plt.show()
```

```python
trump = df.sort_values(by = 'share_voters_voted_trump', ascending = True)
```

```python
trump.head()
```

```python

x=trump['share_voters_voted_trump']
y= trump['avg_hatecrimes_per_100k_fbi']


plt.xlabel('share of voters who voted trump')
plt.ylabel('avg hatecrimes per 100k')
plt.title('Correlation Between % of Trump Votes in States and the Number of Hate Crimes Committed in those States')

#area = (30 * np.random.rand(N))**2  # 0 to 15 point radii

plt.scatter(x, y)
plt.show()
```
